## 1) NLTK: paragraph → sentences

**What it does:** Punkt sentence tokenizer; handles many common abbreviations better than regex.

```python
!pip -q install nltk
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download("punkt", quiet=True)

sentences = sent_tokenize(paragraph)
for i, s in enumerate(sentences, 1):
    print(f"{i}. {s}")
```

## 2) spaCy sentence segmentation

**What it does:** uses a pipeline to segment sentences (can be just a lightweight sentencizer, or full parser).
you can see two modes:
* fast rule-based sentencizer
* more accurate parser-based segmentation

```python
!pip -q install spacy
```

```python
import spacy

print("spaCy sentencizer:")
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")
doc = nlp(paragraph)
for i, s in enumerate(doc.sents, 1):
    print(i, s.text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(paragraph)
sentences = [s.text.strip() for s in doc.sents]

for i, s in enumerate(sentences, 1):
    print(f"{i}. {s}")
```

## 3) Sentence split with regex (baseline)

**What it does:** splits on . ? ! (simple rule).
**Why it’s useful:** shows why naive splitting breaks on abbreviations (U.A.E., Dr., p.m., 2.5).

```python
import re
```

```python
print("A) Regex sentence split:")
sents = [s.strip() for s in re.split(r'(?<=[.!?])\s+', paragraph) if s.strip()]
for i, s in enumerate(sents, 1):
    print(i, s)
```

## Classic NLP Pipeline

```python
import spacy

text = "So excited for NLP with Prof. Elakkiya R at BITS Dubai 😄🔥 #NLPDay1 — BPDC vs BITS Dubai vs BITS Pilani Dubai — don’t/dont/cant/can’t — U.A.E. vs UAE vs United Arab Emirates."
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)
print([t.text for t in doc])
```

### 1) POS tagging (Token → POS)
https://cs.nyu.edu/~grishman/jet/guide/PennPOS.html

```python
print("TOKEN\tPOS\tTAG")
for t in doc:
    print(f"{t.text}\t{t.pos_}\t{t.tag_}")
```

### 2) NER (Named Entity Recognition)
https://learn.microsoft.com/en-us/azure/ai-services/language-service/named-entity-recognition/concepts/named-entity-categories

```python
print("Entities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

### 3) Normalization (typical: lowercase + punctuation removal + unify quotes)

```python
import unicodedata

def normalize_text(s: str) -> str:
    s = unicodedata.normalize("NFKC", s)           # normalize unicode variants
    s = s.replace("’", "'")                        # curly apostrophe → straight
    return s

norm = normalize_text(text)
print(norm)
```

```python
doc_norm = nlp(norm)
print([t.text for t in doc_norm])
```

## 4) Stop words removal (but keep negation words)

Stopword removal is task-dependent. For sentiment, you usually keep not/no/n’t.

```python
stopwords = nlp.Defaults.stop_words

filtered = []
for t in doc_norm:
    if t.is_space:
        continue
    # keep negations even if they appear as stop words
    if t.lower_ in {"not", "no", "n't"}:
        filtered.append(t.text)
        continue
    if t.is_stop:
        continue
    filtered.append(t.text)

print(filtered)
```

### 5) Stemming (Porter stemmer)

spaCy doesn’t stem by default, so use NLTK stemmer.


**stemming is crude (good for search sometimes, very bad for interpretation).**

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

tokens_alpha = [t.text for t in doc_norm if t.is_alpha]   # stem only alphabetic
stems = [ps.stem(tok.lower()) for tok in tokens_alpha]

print(list(zip(tokens_alpha, stems))[:30])
```

## 6) Lemmatization (spaCy lemma)
lemmatization is more linguistically correct than stemming.

```python
nlp = spacy.load("en_core_web_sm")

text = "So excited for NLP with Prof. Elakkiya R at BITS Dubai 😄🔥 #NLPDay1 — BPDC vs BITS Dubai vs BITS Pilani Dubai — don’t/dont/cant/can’t — U.A.E. vs UAE vs United Arab Emirates."
doc = nlp(text)

lemmas = []
for t in doc:
    if t.is_space or t.is_punct:
        continue
    lemmas.append((t.text, t.lemma_))

print("\nLEMMA PAIRS (token -> lemma):")
print(lemmas)
```

```python
text = "studying are is am studied and excited to study"

tokens = word_tokenize(text)                 # tokenize
tokens_alpha = [t for t in tokens if t.isalpha()]   # keep only alphabetic tokens

ps = PorterStemmer()
stems = [ps.stem(tok.lower()) for tok in tokens_alpha]

print("TOKENS (alpha only):")
print(tokens_alpha)

print("\nSTEMS (token -> stem):")
print(list(zip(tokens_alpha, stems)))
```

```python
nlp = spacy.load("en_core_web_sm")

text = "studying are is am studied and excited to study"
doc = nlp(text)

print("TOKENS:")
print([t.text for t in doc])

lemmas = []
for t in doc:
    if t.is_space or t.is_punct:
        continue
    lemmas.append((t.text, t.lemma_))

print("\nLEMMA PAIRS (token -> lemma):")
print(lemmas)
```

## 7) A “clean tokens” pipeline output (what you feed into features)

This gives you a final token list like you’d use for TF-IDF:

```python
clean_tokens = []
for t in doc_norm:
    if t.is_space or t.is_punct:
        continue
    if t.is_stop and t.lower_ not in {"not", "no", "n't"}:
        continue
    # choose lemma (recommended) and lowercase
    tok = t.lemma_.lower()
    clean_tokens.append(tok)

print(clean_tokens)
print("count:", len(clean_tokens))
```
