## 3.1 Predefined Classes
###  \d & \D
* \d returns a match where the string contains digits (numbers from 0-9).
* \D returns a match where the string does not contain any digit. It is basically the opposite of \d.

```python
# \d: Matches any digit
pattern_digit = r"\d"
matches_digit = re.findall(pattern_digit, text)
print("Digits found:", matches_digit)

# \D: Matches any non-digit
pattern_non_digit = r"\D"
matches_non_digit = re.findall(pattern_non_digit, text)
print("Non-digits found:", matches_non_digit)
```

###  \w & \W
* \w helps in extraction of alphanumeric characters only (characters from a to Z, digits from 0-9, and the underscore _ character)
* \W returns match at every non-alphanumeric character. Basically opposite of \w.


```python
# \w: Matches any word character
pattern_word = r"\w"
matches_word = re.findall(pattern_word, text)
print("Word characters found:", matches_word)

# \W: Matches any non-word character
pattern_non_word = r"\W"
matches_non_word = re.findall(pattern_non_word, text)
print("Non-word characters found:", matches_non_word)
```

### \s & \S
* \s: Matches any whitespace
* \S: Matches any non-whitespace

```python
# \s: Matches any whitespace
pattern_whitespace = r"\s"
matches_whitespace = re.findall(pattern_whitespace, text)
print("Whitespaces found:", matches_whitespace)

# \S: Matches any non-whitespace
pattern_non_whitespace = r"\S"
matches_non_whitespace = re.findall(pattern_non_whitespace, text)
print("Non-whitespaces found:", matches_non_whitespace)
```

## 3.4 Lookahead and Lookbehind
Lookahead and lookbehind assertions allow you to match patterns that are followed or preceded by another pattern without including that pattern in the match.

### 3.4.1 Positive Lookahead ((?=...))
Positive Lookahead ((?=...)): Asserts that what follows the current position in the string must match the specified pattern.

```python
import re

text = "foo123, foo456"
pattern = r"foo(?=\d{3})"

matches = re.findall(pattern, text)
print("Positive lookahead matches found:", matches)
```

### 3.4.2 Negative Lookahead ((?!...))
Negative Lookahead ((?!...)): Asserts that what follows the current position in the string must not match the specified pattern.

```python
import re

text = "foo123, fooabc"
pattern = r"foo(?!\d{3})"

matches = re.findall(pattern, text)
print("Negative lookahead matches found:", matches)
```

### 3.4.3 Positive Lookbehind ((?<=...))
Positive Lookbehind ((?<=...)): Asserts that what precedes the current position in the string must match the specified pattern.

```python
import re

text = "123foo, abcfoo"
pattern = r"(?<=\d{3})foo"

matches = re.findall(pattern, text)
print("Positive lookbehind matches found:", matches)
```

### 3.4.4 Negative Lookbehind ((?<!...))
Negative Lookbehind ((?<!...)): Asserts that what precedes the current position in the string must not match the specified pattern.

```python
import re

text = "123foo, abcfoo"
pattern = r"(?<!\d{3})foo"

matches = re.findall(pattern, text)
print("Negative lookbehind matches found:", matches)
```

### **Byte Pair Encoding**

```python
from collections import Counter, defaultdict
import math
import re
```

```python
corpus = {
    "old": 7,
    "older": 3,
    "finest": 9,
    "lowest": 4
}
```

```python
# ----------------------------
# 2) Initialize: split into chars + </w>
#    We'll store each "word" as a tuple of symbols.
# ----------------------------
def init_vocab(word_freqs):
    vocab = {}
    for word, freq in word_freqs.items():
        symbols = tuple(list(word) + ["</w>"])
        vocab[symbols] = vocab.get(symbols, 0) + freq
    return vocab

vocab = init_vocab(corpus)

def pretty_vocab(vocab):
    # prints each token sequence and its frequency
    for sym_seq, freq in sorted(vocab.items(), key=lambda x: (-x[1], x[0])):
        print(f"{freq:>3}  {' '.join(sym_seq)}")

print("=== Initial vocabulary ===")
pretty_vocab(vocab)
```

```python
# ----------------------------
# 3) Count adjacent symbol pairs (weighted)
# ----------------------------
def get_pair_stats(vocab):
    pair_counts = Counter()
    for sym_seq, freq in vocab.items():
        for i in range(len(sym_seq) - 1):
            pair = (sym_seq[i], sym_seq[i+1])
            pair_counts[pair] += freq
    return pair_counts

def print_top_pairs(pair_counts, top_k=20):
    for (a, b), c in pair_counts.most_common(top_k):
        print(f"{(a, b)} : {c}")
```

```python
# ----------------------------
# 4) Merge a chosen pair everywhere in vocab
# ----------------------------
def merge_pair(vocab, pair):
    a, b = pair
    merged = a + b  # like "e" + "s" -> "es"
    new_vocab = {}

    for sym_seq, freq in vocab.items():
        new_seq = []
        i = 0
        while i < len(sym_seq):
            if i < len(sym_seq) - 1 and sym_seq[i] == a and sym_seq[i+1] == b:
                new_seq.append(merged)
                i += 2
            else:
                new_seq.append(sym_seq[i])
                i += 1

        new_seq = tuple(new_seq)
        new_vocab[new_seq] = new_vocab.get(new_seq, 0) + freq

    return new_vocab, merged
```

```python
# ----------------------------
# 5) Train BPE for K merges, showing every step
# ----------------------------
def train_bpe(vocab, num_merges=5, tie_break="first"):
    merges = []          # list of merged pairs in order
    merge_symbols = []   # resulting merged symbols

    for step in range(1, num_merges + 1):
        pair_counts = get_pair_stats(vocab)
        if not pair_counts:
            break

        best_count = pair_counts.most_common(1)[0][1]
        best_pairs = [p for p, c in pair_counts.items() if c == best_count]

        # tie-breaking
        if tie_break == "first":
            # stable: choose lexicographically smallest pair among ties
            best_pair = sorted(best_pairs)[0]
        else:
            best_pair = pair_counts.most_common(1)[0][0]

        print(f"\n=== Merge step {step} ===")
        print("Top pairs (sample):")
        print_top_pairs(pair_counts, top_k=10)
        print(f"\nChosen pair to merge: {best_pair} (count={best_count})")

        vocab, merged_symbol = merge_pair(vocab, best_pair)
        merges.append(best_pair)
        merge_symbols.append(merged_symbol)

        print("\nUpdated vocab:")
        pretty_vocab(vocab)

    return vocab, merges, merge_symbols

# Train for 5 merges (like your example)
trained_vocab, merges, merge_symbols = train_bpe(vocab, num_merges=5, tie_break="first")

print("\n=== Learned merges ===")
for i, (p, ms) in enumerate(zip(merges, merge_symbols), start=1):
    print(f"{i:>2}. {p[0]} + {p[1]} -> {ms}")
```

```python
# ----------------------------
# 6) Build a subword "tokenizer" from merges (apply merges in order)
#    This encodes any new word into subword tokens.
# ----------------------------
def encode_with_bpe(word, merges):
    symbols = list(word) + ["</w>"]
    # Apply merges in order
    for a, b in merges:
        i = 0
        out = []
        while i < len(symbols):
            if i < len(symbols) - 1 and symbols[i] == a and symbols[i+1] == b:
                out.append(a + b)
                i += 2
            else:
                out.append(symbols[i])
                i += 1
        symbols = out
    return symbols

test_words = ["old", "older", "finest", "lowest", "oldest"]
print("\n=== Encoding examples using learned merges ===")
for w in test_words:
    print(w, "->", encode_with_bpe(w, merges))
```

```python
# ----------------------------
# 7) Subword-based retrieval (simple TF-IDF)
#    Docs are your words; query is any string.
# ----------------------------
docs = ["old", "older", "finest", "lowest"]
doc_freqs = corpus  # use their given frequencies if you like; retrieval often treats each doc once

def tokenize_doc(word):
    toks = encode_with_bpe(word, merges)
    # remove </w> marker for retrieval
    return [t for t in toks if t != "</w>"]

doc_tokens = {d: tokenize_doc(d) for d in docs}

print("\n=== Document tokens (subwords) ===")
for d in docs:
    print(d, "->", doc_tokens[d])

def compute_idf(doc_tokens_dict):
    N = len(doc_tokens_dict)
    df = Counter()
    for d, toks in doc_tokens_dict.items():
        for t in set(toks):
            df[t] += 1
    # smoothed IDF
    idf = {t: math.log((N + 1) / (df_t + 1)) + 1.0 for t, df_t in df.items()}
    return idf

idf = compute_idf(doc_tokens)

def tf_idf_score(query, doc, doc_tokens, idf):
    q_toks = tokenize_doc(query)
    d_toks = doc_tokens[doc]
    tf = Counter(d_toks)
    score = 0.0
    for t in q_toks:
        score += tf[t] * idf.get(t, 0.0)
    return score, q_toks

def retrieve(query, top_k=4):
    scores = []
    for d in docs:
        s, q_toks = tf_idf_score(query, d, doc_tokens, idf)
        scores.append((s, d))
    scores.sort(reverse=True)
    return q_toks, scores[:top_k]

query = "oldest"
q_toks, ranked = retrieve(query)

print("\n=== Retrieval with subwords ===")
print("Query:", query)
print("Query tokens:", q_toks)
print("Ranking (TF-IDF overlap):")
for s, d in ranked:
    print(f"{d:>7}  score={s:.4f}  doc_tokens={doc_tokens[d]}")
```

### **WordPiece Tokenizer**

```python
from collections import Counter
from typing import Dict, List, Tuple

# ----------------------------
# 1) Input corpus (word -> frequency)
# ----------------------------
CORPUS: Dict[str, int] = {"old": 7, "older": 3, "finest": 9, "lowest": 4}
```

```python
# ----------------------------
# 2) Helpers: init, longest-match tokenization
# ----------------------------
def init_tokenizations(corpus: Dict[str, int]) -> Dict[str, List[str]]:
    """
    Initialize each word as characters + </w>.
    """
    tok = {}
    for w in corpus:
        tok[w] = list(w) + ["</w>"]
    return tok


def build_vocab_from_tokenizations(tokenizations: Dict[str, List[str]]) -> set:
    """
    Current vocab is all unique tokens appearing in tokenizations.
    """
    vocab = set()
    for toks in tokenizations.values():
        vocab.update(toks)
    return vocab


def longest_match_tokenize(word: str, vocab: set) -> List[str]:
    """
    Longest-match-first tokenization with current vocab.
    Uses </w> marker: we tokenize word + </w> as one stream.
    """
    symbols = word + "</w>"  # treat </w> as literal suffix in matching
    # But we want tokens like "</w>" to be a symbol.
    # We'll do matching over a string, with special handling for "</w>".
    # Simpler: tokenize over a list of characters + ["</w>"], but allow
    # multi-char tokens within the character stream (excluding </w>).
    stream = list(word) + ["</w>"]

    # Create a fast lookup for candidate token lengths (excluding </w>)
    # Tokens may be like "fin", "st", "old", and also "</w>".
    max_len = max([len(t) for t in vocab if t != "</w>"] + [1])

    out = []
    i = 0
    while i < len(stream):
        # If we're at </w>, it must match exactly
        if stream[i] == "</w>":
            out.append("</w>")
            i += 1
            continue

        best = None
        # Try longest token first
        for L in range(min(max_len, len(stream) - i), 0, -1):
            # Stop if the span would include </w>
            if "</w>" in stream[i : i + L]:
                continue
            cand = "".join(stream[i : i + L])
            if cand in vocab:
                best = cand
                break

        if best is None:
            # Fallback: single character must be in vocab
            best = stream[i]
        out.append(best)
        i += len(best)

    return out
```

```python
# ----------------------------
# 3) Counting: token counts and adjacent pair counts (weighted)
# ----------------------------
def compute_counts(
    tokenizations: Dict[str, List[str]], corpus_freq: Dict[str, int]
) -> Tuple[Counter, Counter]:
    token_counts = Counter()
    pair_counts = Counter()

    for w, toks in tokenizations.items():
        f = corpus_freq[w]
        for t in toks:
            token_counts[t] += f
        for i in range(len(toks) - 1):
            pair_counts[(toks[i], toks[i + 1])] += f

    return token_counts, pair_counts


def score_pair(pair: Tuple[str, str], token_counts: Counter, pair_counts: Counter) -> float:
    a, b = pair
    cab = pair_counts[pair]
    ca = token_counts[a]
    cb = token_counts[b]
    if cab == 0 or ca == 0 or cb == 0:
        return 0.0
    return cab / (ca * cb)
```

```python
# ----------------------------
# 4) WordPiece training loop
# ----------------------------
def train_wordpiece(
    corpus_freq: Dict[str, int],
    num_iters: int = 10,
    print_top_k: int = 12,
) -> Tuple[Dict[str, List[str]], List[Tuple[str, str]]]:
    tokenizations = init_tokenizations(corpus_freq)
    vocab = build_vocab_from_tokenizations(tokenizations)
    merges: List[Tuple[str, str]] = []

    print("=== Initial tokenizations ===")
    for w in corpus_freq:
        print(f"{w:>7} ({corpus_freq[w]}): {' '.join(tokenizations[w])}")
    print("\n=== Initial vocab ===")
    print(sorted(vocab))

    for step in range(1, num_iters + 1):
        token_counts, pair_counts = compute_counts(tokenizations, corpus_freq)

        if not pair_counts:
            print("\nNo more pairs to merge.")
            break

        # Compute scores for all pairs
        scored = []
        for p in pair_counts:
            s = score_pair(p, token_counts, pair_counts)
            scored.append((s, p, pair_counts[p]))

        scored.sort(key=lambda x: (-x[0], -x[2], x[1]))  # score desc, then count desc, then lex

        best_score, best_pair, best_count = scored[0]
        if best_score == 0.0:
            print("\nAll remaining pairs have 0 score; stopping.")
            break

        a, b = best_pair
        new_tok = a + b
        merges.append(best_pair)

        print(f"\n\n==============================")
        print(f"Step {step}")
        print(f"==============================")
        print("Top candidate pairs (score, pair, pair_count, count(a), count(b)):")
        for s, p, c in scored[:print_top_k]:
            aa, bb = p
            print(
                f"{s:.6f}  {p!s:>18}  pair_count={c:<3}  "
                f"count({aa})={token_counts[aa]:<3}  count({bb})={token_counts[bb]:<3}"
            )

        print(f"\nChosen merge: {best_pair}  ->  '{new_tok}'")
        print(f"score={best_score:.6f}, pair_count={best_count}")

        # Add new token to vocab
        vocab.add(new_tok)

        # Re-tokenize all words using longest-match-first
        for w in corpus_freq:
            tokenizations[w] = longest_match_tokenize(w, vocab)

        print("\nUpdated tokenizations:")
        for w in corpus_freq:
            print(f"{w:>7} ({corpus_freq[w]}): {' '.join(tokenizations[w])}")

        print("\nCurrent vocab (sorted):")
        print(sorted(vocab))

    return tokenizations, merges


if __name__ == "__main__":
    final_tokenizations, merges = train_wordpiece(CORPUS, num_iters=10, print_top_k=12)

    print("\n\n=== Final merges learned (in order) ===")
    for i, (a, b) in enumerate(merges, start=1):
        print(f"{i:>2}. {a} + {b} -> {a+b}")

    print("\n=== Final tokenizations ===")
    for w, toks in final_tokenizations.items():
        print(f"{w:>7}: {' '.join(toks)}")
```

```python
# Compare tokenization on the SAME corpus words using:
# (1) GPT-2 (byte-level BPE)  and  (2) BERT (WordPiece)
#
# pip install transformers

from transformers import GPT2TokenizerFast, BertTokenizerFast

words = ["old", "older", "finest", "lowest"]

# GPT-2 tokenizer (byte-level BPE)
gpt_tok = GPT2TokenizerFast.from_pretrained("gpt2")

# BERT tokenizer (WordPiece)
bert_tok = BertTokenizerFast.from_pretrained("bert-base-uncased")

def show_gpt2(word: str):
    # GPT-2 is very sensitive to whitespace.
    # In real text, tokens often come with a leading space for non-initial words.
    no_space_tokens = gpt_tok.tokenize(word)
    space_tokens = gpt_tok.tokenize(" " + word)

    no_space_ids = gpt_tok.encode(word, add_special_tokens=False)
    space_ids = gpt_tok.encode(" " + word, add_special_tokens=False)

    print(f"GPT-2 word: '{word}'")
    print(f"  tokenize('{word}')      -> {no_space_tokens} | ids={no_space_ids}")
    print(f"  tokenize(' {word}')     -> {space_tokens} | ids={space_ids}")
    print(f"  decode(ids no-space)    -> {repr(gpt_tok.decode(no_space_ids))}")
    print(f"  decode(ids with-space)  -> {repr(gpt_tok.decode(space_ids))}")

def show_bert(word: str):
    # BERT WordPiece uses '##' prefix for subword continuations
    tokens = bert_tok.tokenize(word)
    ids = bert_tok.encode(word, add_special_tokens=True)  # includes [CLS] and [SEP]
    ids_no_special = bert_tok.encode(word, add_special_tokens=False)

    print(f"BERT word: '{word}'")
    print(f"  tokenize('{word}')              -> {tokens}")
    print(f"  encode(add_special_tokens=True) -> {ids} | decoded={bert_tok.decode(ids)}")
    print(f"  encode(add_special_tokens=False)-> {ids_no_special} | decoded={bert_tok.decode(ids_no_special)}")

print("========== GPT-2 (byte-level BPE) ==========")
for w in words:
    show_gpt2(w)
    print()

print("\n========== BERT (WordPiece) ==========")
for w in words:
    show_bert(w)
    print()

# Optional: show a query-like word that wasn't in your corpus list (to see subword behavior)
extra = "oldest"
print("\n========== EXTRA: 'oldest' ==========")
show_gpt2(extra)
print()
show_bert(extra)
```
